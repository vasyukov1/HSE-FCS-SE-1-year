## Применение интегралов для вычисления площади фигуры и длины дуги

### Задание 1
Вычислить площадь фигуры, ограниченной графиками функций $f = 2 + x^2$ и  $g = 2 - x^3$. Построить графики, найти точки пересечения, затем вычислить соответствующие интегралы.

```python
from sympy.abc import x

f = 2 + x**2
g = 2 - x**3

plot(f, g, (x, -2, 2), ylim=(-1, 4))
roots = solveset(sympy.Eq(f, g), domain=S.Reals)
area_fg = (f - g).integrate((x, min(roots), max(roots)))

display(Latex(fr"""\text{{Roots: }} {latex(roots)}"""))
display(Latex(fr"""\text{{Area: }} {latex(area_fg)}"""))
```

### Задача 2.
Вычислить площадь фигуры, ограниченной графиком функции $x = \cos^3(t)$, $y = \sin^3(t)$, $0 \le t \le 2\pi$.

```python
from sympy.abc import t

x = Cos(t)**3
y = Sin(t)**3

area_xy = (x * y.diff(t)).integrate((t, 0, 2*Pi))

display(Latex(fr"""\text{{Area: }} {latex(area_xy)}"""))
```

### Задача 3.
Вычислить площадь фигуры, ограниченной графиком функции $r = \cos(2\phi)$, $-\frac{\pi}{4} \le \phi \le \frac{\pi}{4}$.

```python
phi = S('phi')
r = Cos(2*phi)

area_r = S(1)/2 * integrate(r**2, (phi, -Pi/4, Pi/4))

display(Latex(fr"""\text{{Area: }} {latex(area_r)}"""))
```

### Задача 4.
Вычислить длину кривой, заданной уравнением $y^2 = x^3$ при $x$ от 1 до 3.

```python
from sympy.abc import x

y = x ** (S(3) / 2)

# Умножаем на 2, так как |y|, а не просто y при y^2 = x^3.
length_y = 2 * Curve((x, x** (S(3/2))), (x, 1, 3)).length

display(Latex(fr"""\text{{Length: }} {latex(length_y)}"""))
```

### Задача 5.
Вычислить длину кривой, заданной параметрически $x = \frac{3}{2}\cos^3 t $, $y = 3\sin^3 t$.

```python
from sympy.abc import t

x = S(3)/2 * Cos(t) ** 3
y = 3 * Sin(t) ** 3

length_xy = sympy.simplify(integrate((x.diff(t)**2 + y.diff(t)**2) ** S.Half, (t, 0, 2 * Pi)))

display(Latex(fr"""\text{{Length: }} {latex(length_xy)}"""))
```

### Задача 6.
Вычислить длину кривой, заданной в полярных координатах
$r = \phi$, $\phi$ от 0 до $2\pi$. Результат вычислить приближенно до 3 знаков после запятой.

```python
from sympy.abc import phi

r = phi
length_phi = round(((r**2 + r.diff(phi)**2)**S.Half).integrate((phi, 0, 2*Pi)), 3)

display(Latex(fr"""\text{{Length: }} {latex(length_phi)}"""))
```

### Индивидуальное задание
Вычислить площадь фигуры, ограниченной графиками функций $f = 2\sin(x)$ и  $g = \cos(2x)$ на интервале  $[0, \pi]$. Построить графики, найти точки пересечения (выделить корни на интервале с помощью intersect и Interval), затем вычислить соответствующие интегралы.

image.png

```python
from sympy.abc import x

f = 2 * Cos(2*x)
g = 5 * Cos(x)

plot(f, g, (x, -2, 6), ylim=(-6, 6))

roots = solveset(sympy.Eq(f, g), domain=S.Reals).intersect(Interval(0.1*Pi, 1.9*Pi))
area = sympy.simplify((f - g).integrate((x, min(roots), max(roots))))

display(Latex(fr"""\text{{Roots: }} {latex(roots)}"""))
display(Latex(fr"""\text{{Area: }} {latex(area)}"""))
```
