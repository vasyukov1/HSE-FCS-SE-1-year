### Задание 1
Решить уравнение $6x^4 + 9x^3 - 99x^2 - 66x + 360 = 0$ с помощью solve и solveset.

```python
x = Symbol('x')
print(f"solve: {solve(6*x**4 + 9*x**3 - 99*x**2 - 66*x + 360)}")
print(f"solveset: {solveset(6*x**4 + 9*x**3 - 99*x**2 - 66*x + 360)}")
```

### Задание 6.
Найти решение уравнения $\frac{(x^3 + 5 x^2 - 3 x - 15)\ln(x^2 - 3)}{x^2 - 3 x + 2} = 0$. Выполнить проверку подстановкой subs.
#### Указание.
Решение нужно разбить на части. Нужно учесть область определения выражения в левой части уравнения.

```python
equation = (x**3 + 5*x**2 - 3*x - 15) * sympy.ln(x**2 - 3) / (x**2 - 3*x + 2)
res = solveset(equation)
print(f"result: {res}")

print("-----check-----")
for i in res:
  print(f"f({i}) = {equation.subs(x, i)}")
```

### Индивидуальное задание.
Через все точки пересечения графиков функций $f(x)$ и $g(x)$ провести горизонтальные прямые. Изобразить на графике $f(x)$ зеленым цветом, $g(x)$ красным, горизонтальные линии черным.
Формулы для $f(x)$ и $g(x)$ включить в легенду.

image.png

```python
f = -(x**2 + 15*x + 84)/12
g = x**2/2 + 4*x + 7/2

f_graph = sympy.plot(f, line_color="green", legend=True, show=False)
g_graph = sympy.plot(g, line_color="red", legend=True, show=False)

res = sympy.plot(ylim=(-10, 10), show=False)
res.extend(f_graph)
res.extend(g_graph)

dots = solveset(f - g)
for i in dots:
  new_line = f.subs(x, i)
  res.extend(sympy.plot(new_line, line_color="black", show=False))

res.show()
```
