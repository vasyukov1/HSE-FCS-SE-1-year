```python
import numpy as np
import sympy
from sympy import nsolve
import matplotlib.pyplot as plt
from IPython.display import display, Latex
%matplotlib inline
```

### –ó–∞–¥–∞—á–∞ 1
–†–µ—à–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ $ùë•^4‚àíe^x=0$ –ø—Ä–∏ $-2 \le x \le 2$, –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫ –ª–µ–≤–æ–π —á–∞—Å—Ç–∏, –æ—Ç–º–µ—Ç–∏—Ç—å —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è —Å –æ—Å—å—é –û–•.

```python
def f(x, lib = 'sympy'):
  if lib == 'sympy':
    return x**4 - sympy.exp(x)
  if lib == 'numpy':
    return x**4 - np.exp(x)
  return '–ö–ª–æ—É–Ω, –ø–∏—à–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ!'

x = sympy.Symbol('x')
X = np.linspace(-2, 2)

roots = [sympy.nsolve(sympy.Eq(f(x, lib='sympy'), 0), x, x0) for x0 in [-2, 2]]
display(Latex(f'\\text{{–ö–æ—Ä–Ω–∏ —É—Ä–∞–≤–Ω–µ–Ω–∏—è: }}{sympy.latex(roots)}'))

ax = plt.gca()
ax.plot(X, f(X, lib='numpy'), color='pink')
y_coord = [f(float(root), lib='numpy') for root in roots]
ax.scatter(roots, y_coord)
for i, x_coord in enumerate(roots):
  ax.annotate('X' + str(i + 1), xy = (x_coord, y_coord[i]), xycoords= ' data',
              xytext = (x_coord - 0.1, y_coord[i] + 0.3),  textcoords='data')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
ax.plot(1, 0, '>k', transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, '^k', transform=ax.get_xaxis_transform(), clip_on=False)
```

### –ó–∞–¥–∞—á–∞ 2.
–†–µ—à–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ $\sin^2(2x)=\ln x$ –ø—Ä–∏ $0< x \le 3$. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –ª–µ–≤–æ–π –∏ –ø—Ä–∞–≤–æ–π —á–∞—Å—Ç–∏, –æ—Ç–º–µ—Ç–∏—Ç—å —Ç–æ—á–∫—É –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤.

```python
def f(x, lib = 'sympy'):
  if lib == 'sympy':
    return (sympy.sin(2 * x)) ** 2
  if lib == 'numpy':
    return (np.sin(2 * x)) ** 2
  return '–ö–ª–æ—É–Ω, –ø–∏—à–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ!'

def g(x, lib = 'sympy'):
  if lib == 'sympy':
    return sympy.ln(x)
  if lib == 'numpy':
    return np.log(x)
  return '–ö–ª–æ—É–Ω, –ø–∏—à–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ!'

x = sympy.Symbol('x')
X = np.linspace(10**(-2), 3)

roots = [sympy.nsolve(sympy.Eq(f(x, lib='sympy'), g(x, lib='sympy')), x, x0) for x0 in [10**(-5), 2.7, 3.0]]
display(Latex(f'\\text{{–ö–æ—Ä–Ω–∏ —É—Ä–∞–≤–Ω–µ–Ω–∏—è: }}{sympy.latex(roots)}'))

ax = plt.gca()
ax.plot(X, f(X, lib='numpy'), color='blue')
ax.plot(X, g(X, lib='numpy'), color='red')
y_coord = [f(float(root), lib='numpy') for root in roots]
ax.scatter(roots, y_coord)
for i, x_coord in enumerate(roots):
  ax.annotate('X' + str(i + 1), xy = (x_coord, y_coord[i]), xycoords= ' data',
              xytext = (x_coord - 0.1, y_coord[i] + 0.3),  textcoords='data')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
ax.plot(1, 0, '>k', transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, '^k', transform=ax.get_xaxis_transform(), clip_on=False)
```

### –ó–∞–¥–∞—á–∞ 3.
–†–µ—à–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ $\sin^2(2x^2-7x+6)=3/4$ –Ω–∞ $[3, 4]$. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –ª–µ–≤–æ–π –∏ –ø—Ä–∞–≤–æ–π —á–∞—Å—Ç–∏, –æ—Ç–º–µ—Ç–∏—Ç—å —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤.

```python
def f(x, lib = 'sympy'):
  if lib == 'sympy':
    return (sympy.sin(2*x**2 - 7*x + 6)) ** 2
  if lib == 'numpy':
    return (np.sin(2*x**2 - 7*x + 6)) ** 2
  return '–ö–ª–æ—É–Ω, –ø–∏—à–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ!'

def g(x):
  return 3/4 + 0 * x

x = sympy.Symbol('x')
X = np.linspace(3, 4)

roots = [sympy.nsolve(sympy.Eq(f(x, lib='sympy'), g(x)), x, x0) for x0 in [3, 3.1, 3.5, 3.7]]
display(Latex(f'\\text{{–ö–æ—Ä–Ω–∏ —É—Ä–∞–≤–Ω–µ–Ω–∏—è: }}{sympy.latex(roots)}'))

ax = plt.gca()
ax.plot(X, f(X, lib='numpy'), color='blue')
ax.plot(X, g(X), color='red')
y_coord = [f(float(root), lib='numpy') for root in roots]
ax.scatter(roots, y_coord)
for i, x_coord in enumerate(roots):
  ax.annotate('X' + str(i + 1), xy = (x_coord, y_coord[i]), xycoords= ' data',
              xytext = (x_coord - 0.01, y_coord[i] + 0.03),  textcoords='data')

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
```

### –ó–∞–¥–∞—á–∞ 4.
–†–µ—à–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ ${\rm tg}^4(2x^2+1)-10=0$ –Ω–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ $[0, 0.2]$ –º–µ—Ç–æ–¥–æ–º –¥–∏—Ö–æ—Ç–æ–º–∏–∏. –ì—Ä–∞—Ñ–∏–∫ —Å—Ç—Ä–æ–∏—Ç—å –Ω–µ –Ω—É–∂–Ω–æ.

```python
def f(x, lib='sympy'):
    if lib == 'sympy':
        return (sympy.tan(2*x**2 + 1)) ** 4 - 10
    if lib == 'numpy':
        return (np.tan(2*x**2 + 1)) ** 4 - 10
    return '–ö–ª–æ—É–Ω, –ø–∏—à–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ!'

roots = [sympy.nsolve(sympy.Eq(f(x, lib='sympy'), 0), x, interval, verify=True) for interval in [[0, 0.1], [0.1, 0.2]]]
display(Latex(f'\\text{{–ö–æ—Ä–Ω–∏ —É—Ä–∞–≤–Ω–µ–Ω–∏—è: }}{sympy.latex(roots)}'))
```

### –ó–∞–¥–∞—á–∞ 5.
–†–µ—à–∏—Ç—å —Å–∏—Å—Ç–µ–º—É —É—Ä–∞–≤–Ω–µ–Ω–∏–π
$$
\left\{\begin{matrix}
(x - 1)^2 + (y - 2)^2 + (z + 3)^2 = 25\\
x + y + z = 0\\
-2x + 3y - z = 7
\end{matrix}\right.
$$
–Ω–∞—á–∞–ª—å–Ω–æ–µ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ (0, 0, 0).
–ü—Ä–æ–≤–µ—Ä–∏—Ç—å –ø–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–æ–π.

```python
x, y, z = sympy.symbols('x y z')

def f(x, y, z):
    return (x - 1)**2 + (y - 2)**2 + (z + 3)**2

def g(x, y, z):
    return x + y + z

def h(x, y, z):
    return -2*x + 3*y - z

root = sympy.nsolve((sympy.Eq(f(x, y, z), 25), sympy.Eq(g(x, y, z), 0), sympy.Eq(h(x, y, z), 7)), (x, y, z), (0, 0, 0))
display(Latex(f'\\text{{–ö–æ—Ä–Ω–∏ —É—Ä–∞–≤–Ω–µ–Ω–∏—è: }}{sympy.latex(root)}'))

x0, y0, z0 = root
check = f(x0, y0, z0) == 25 and round(g(x0, y0, z0)) == 0 and h(x0, y0, z0) == 7
display(Latex(f'\\text{{–ü—Ä–æ–≤–µ—Ä–∫–∞: }}{sympy.latex(check)}'))
```

### –ò–Ω–¥–∏–≤–∏–¥—É–∞–ª—å–Ω–æ–µ –∑–∞–¥–∞–Ω–∏–µ
–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ —Ñ—É–Ω–∫—Ü–∏–π $f(x)$ –∏ $g(x)$, —É—Ç–æ—á–Ω–∏—Ç—å –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã —Ç–æ—á–µ–∫ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è, —Ä–µ—à–∞—è —á–∏—Å–ª–µ–Ω–Ω–æ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ. –ù–∞  –≥—Ä–∞—Ñ–∏–∫–µ –æ—Ç–º–µ—Ç–∏—Ç—å –∏ –ø–æ–¥–ø–∏—Å–∞—Ç—å –±—É–∫–≤–∞–º–∏ $–ê_1$, $A_2$, ... —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤.  

–í–∞—Ä–∏–∞–Ω—Ç $121$
\begin{align*}
    f(x) = \frac{x^{2}}{2} && g(x) = - \frac{\left(x + 4\right) \left(5 x - 12\right)}{6}
\end{align*}


```python
def f(x):
    return x**2 / 2
def g(x):
    return -(x+4) * (5*x -12) / 6

X = np.linspace(-7, 7)
roots = [sympy.nsolve(sympy.Eq(f(x), g(x)), x, x0) for x0 in [-10, 10]]

ax = plt.gca()
ax.plot(X, f(X), label=r'f(x)', color='blue')
ax.plot(X, g(X), label=r'g(x)', color='red')
y_coord = [f(float(root)) for root in roots]
ax.scatter(roots, y_coord)
for i, x_coord in enumerate(roots):
    ax.annotate('A' + str(i + 1), xy=(x_coord, y_coord[i]), xycoords='data',
                xytext=(x_coord - 0.2, y_coord[i] + 2), textcoords='data')

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
ax.legend(loc='best')
```
